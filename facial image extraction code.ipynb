```python
import cv2

#얼굴 크롭 함수
def cutting_face_save(image):
    face_cascade = cv2.CascadeClassifier("/Users/minseyeon/성균관대학교/3학년 1학기/URP(박은일)/CV/CODE/haarcascade_frontalface_default.xml")
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    cropped_faces = []
    for (x, y, w, h) in faces:
        cropped = image[y:y + h, x:x + w]
        resized = cv2.resize(cropped, (256, 256))
        cropped_faces.append(resized)
    return cropped_faces
```

```python
#비디오 불러오기
video_path = f"/Users/minseyeon/성균관대학교/3학년 1학기/URP(박은일)/CV/video/video64.mp4"
video = cv2.VideoCapture(video_path)

#프레임 캡쳐
cnt=0
ret=True
while(ret):
    ret, frame = video.read()
    cropped_faces=cutting_face_save(frame)
    if int(video.get(1))%20==0:
        for idx, face in enumerate(cropped_faces):
            file_path=f'/Users/minseyeon/성균관대학교/3학년 1학기/URP(박은일)/CV/face_cut/video64_{cnt}.jpg'
            save_success=cv2.imwrite(file_path, face)
            if save_success:
                print(f'Image_{cnt} saved')
            else:
                print(f'Failed to save Image_{cnt}')
        cnt += 1

    if ret==False:
        break

#비디오 파일 닫기    
video.release()
```
