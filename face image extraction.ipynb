```python
######################
# 얼굴 인식 함수 설정#
######################

import cv2
import os

# 얼굴 인식 및 분리 함수 설정
def cutting_face_save(image):
    # 얼굴 인식 함수 : OpenCV에서 제공하는 사전 훈련된 정면 인식 함수 설정
    face_cascade = cv2.CascadeClassifier("C:/Users/ehdal/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml")

    # 얼굴 인식 정확도를 높히기 위해 grayscale 설정
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 기준에 맞게 얼굴 검출
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # 얼굴 이미지 저장할 리스트 생성
    cropped_faces = []

    # 검출된 얼굴에 대해 반복문 설정 : 사이즈 재조정 및 저장
    for (x, y, w, h) in faces:
        # 얼굴 이미지 저장
        cropped = image[y:y + h, x:x + w]
        # 얼굴 이미지를 256x256 사이즈로 재조정
        resized = cv2.resize(cropped, (256, 256))
        # 조정된 이미지를 얼굴 리스트에 저장
        cropped_faces.append(resized)

    return cropped_faces
```

```python
##################
# 얼굴 인식 시행 #
##################

# 저장할 이미지 경로 및 비디오 파일 형식 설정
dataset_path = 'C:/Users/ehdal/Documents/Bandicam/sober_mp4_DIF'
video_files = [f for f in os.listdir(dataset_path) if f.endswith('.mp4')]

# 각 비디오 파일에 대한 반복문 실행
for video_file in video_files:
    video_path = os.path.join(dataset_path, video_file)
    video = cv2.VideoCapture(video_path)

    # 이미지 캡쳐 간격을 20프레임으로 설정
    frame_interval = 20

    # 이미지 이름 구분을 위한 카운트 설정
    face_count = 1

    # 반복문을 위한 임시 변수 설정
    ret = True

    while(ret):
        # 비디오 인식 후 프레임 추출
        ret, frame = video.read()
        frame_number = int(video.get(cv2.CAP_PROP_POS_FRAMES))

        # 인식이 성공했고, 추출 프레임에 해당할 경우 반복문 시행
        if ret and frame_number % frame_interval == 0:
            cropped_faces = cutting_face_save(frame)
            for face in cropped_faces:
                # 저장할 이미지 파일의 이름 설정.
                face_image_name = f'{video_file[:-4]}_{face_count}.jpg'
                # 이미지 저장 경로 설정
                face_image_path = os.path.join('C:/Users/ehdal/Desktop/dataset/sober_image_layer2_1123/', face_image_name)
                # 이미지 파일을 저장 후 저장 완료 메세지 출력
                save_success = cv2.imwrite(face_image_path, face)
                if save_success:
                    print(f'Image saved: {face_image_name}')
                else:
                    print(f'Failed to save image: {face_image_name}')
                # 다음 이미지 이름 설정 (+ 1)
                face_count += 1

    # 비디오 처리 완료
    video.release()
```
